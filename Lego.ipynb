<h3>Understanding LEGO

Themes vs. LEGO Sets</h3>


When you enter a LEGO store, you will discover that their merchandise is arranged by theme. Star Wars, Batman, Harry Potter, and many other things are among their themes.


*A* lego set is a particular box of LEGO or product. Therefore, a single theme typically has many different sets.

<img src="https://i.imgur.com/whB1olq.png">

```python
df_set = pd.read_csv("data/sets.csv")
df_set
```

A different data collection called sets provides additional details about Lego items, such as the names of the sets that were launched, the year they were released, and the total number of bricks or pieces they contained. This undoubtedly expands the range of questions that may be answered and the amount of data that can be gleaned from it.

```python
first_sets = df_set.sort_values(by=['year'], ascending=True)
first_sets.head(5)
```

Now that we are aware that the first Lego sets date back to 1949, when they were first released, we can say that they had some generic titles but none of the specific themes that Lego is known for today.

```python
different_sets = df_set.loc[df_set['year'] == 1949]
different_sets
```

In their first year of operation in 1949, LEGO had a decent number, or five different types of sets, available for purchase, of which 4 were complete sets and the fifth merely included supplementary or supplemental components for the other four sets.


```python
biggest_set = df_set.sort_values(by=['num_parts'], ascending=False)
biggest_set.head(5)
```

Up to this point, Lego has produced an impressive amount of sets with a lot of bricks and a variety of themes. However the position for "The set with the highest number of parts" unquestionably belongs to "The ultimate numbre of Chimera," which features a whopping number of 9987 pieces, or nearly ten thousand pieces in a single set.

```python
year_on_year = df_set.groupby(['year']).count()
year_on_year['set_num']
```

This demonstrates that between the starting year of 1949 and the year 2021, the number of sets on sale surely increased enormously.

```python
old_year = df_set.loc[df_set['year'] == 1955].count()
latest_year = df_set.loc[df_set['year'] == 2019].count()
old_year
latest_year
```

```python
plt.plot(year_on_year.index[:-2], year_on_year.set_num[:-2], color="silver", linewidth="5")
```

This gives us a visual represntation of the sets released by lego right from 1950 to 2020 which is equal to the span of 70 years

```python
themes = pd.read_csv("data/themes.csv")
themes
```

Another Lego Company data set called themes can assist us comprehend the many themes around which the company created new sets.

```python
themes_by_year = df_set.groupby('year').agg({'theme_id': pd.Series.nunique})
```

```python
themes_by_year.rename(columns = {'theme_id':'nr_themes'},inplace=True)
themes_by_year.head(5)
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], color="pink", linewidth="5")
```

This demonstrates without a doubt that as the number of sets shot up in sales, the number of themes for the Lego sets also increased in line with this pattern and saw a significant increase in the number of themes made available to its customers.


### Line Charts with Two Seperate Axes

```python
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.plot(year_on_year.index[:-2], year_on_year.set_num[:-2], color='silver', linewidth="5")
ax2.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], color='pink', linewidth="5")
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of sets')
ax2.set_ylabel('Number of themes')
```

This makes it simpler for us to comprehend how well the number of sets and the themes scaled up throughout the years, and it is undeniable that Lego has undergone many beneficial improvements in its products, whether it be the number of sets or the number of themes released year over year.


```python
parts_per_set = df_set.groupby('year').agg({'num_parts': pd.Series.mean})
parts_per_set
```

The number of parts in the lego sets has increased by a fair but not an overwhelming amount with the average number of parts in each set ranging from 50 to to around 300.

### Scatter Plots

```python
plt.scatter(parts_per_set.index[:-2], parts_per_set.num_parts[:-2], s = 100, alpha = 0.6, cmap = 'nipy_spectral', color='grey')
```

The number of parts over the years has seen a fair amount of increase with the older products having around 50 parts to the latest products having as much as 250 to 300 parts per set.

### Merging DataFrames based on a Key


```python
set_theme_count = pd.DataFrame({'id':set_theme_count.index,'set_count':set_theme_count.values})
set_theme_count.head(5)
```

```python
df_merged = pd.merge(set_theme_count, themes, on='id')
df_merged
```

```python
plt.figure(figsize=(12,6))
plt.xticks(fontsize=14, rotation=40)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)
plt.bar(df_merged.name[:10], df_merged.set_count[:10],color=['silver','pink'])
```

This shows us some of the most popular products based on themes and it is evident that star wars won by a large margin. However Lego did not restrict itself to only brick toys and has expanded over a wide range of products like key chains and book wich are a quite a seller as well
